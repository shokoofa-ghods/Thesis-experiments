```python
import shutil

def move_previous_data(path):
    local = origin + 'K-fold Experiments (copy)/fold 2/Train/' 
    source_dir = os.path.join(local, path)
    files = os.listdir(source_dir)

    dest = os.path.join(local, path, '0')
    print(dest)
    os.makedirs(dest, exist_ok=True)

    for file in files:
        src = os.path.join(local + path, file)
        dst = os.path.join(dest, file)

        if os.path.isfile(src):
            shutil.move(src, dest)
    
```

```python
def create_2D_images(img, assigned_Rpeaks_list, updated_path):
    local = origin + 'K-fold Experiments (copy)/fold 2/Train/' 
    dir = os.path.join(local, updated_path)
    for i in range(len(assigned_Rpeaks_list)-1):
        R_start = assigned_Rpeaks_list[i]
        R_end = assigned_Rpeaks_list[i+1]

        dest = os.path.join(dir, f'{i+1}')
        os.makedirs( dest, exist_ok=True)
        print(dest)
        
        file_name= updated_path.split('/')[-1]

        for idx in range(R_start, R_end+1):
            im = Image.fromarray(img[idx,:,:])
            im.save(os.path.join( dest, (f"{file_name}_{idx}.jpeg"))) 
```

```python
for idx, path in enumerate(r['name']):
    path = '/'.join(path.split(' '))
    move_previous_data(path)
```

```python
for idx, path in enumerate(r['name']):
    path = '/'.join(path.split(' '))
    source_path = os.path.join(origin + 'Tahseen/Train', path)
    dcm = pydicom.dcmread(source_path)
    img = dcm.pixel_array
    peaks = r.iloc[idx, 2]
    create_2D_images(img[:,:,:,0], peaks, path)

```
