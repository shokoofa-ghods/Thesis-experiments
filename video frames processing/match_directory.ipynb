```python
import shutil 

def move_files(df, source_dir, set_data):
    for index, row in df.iterrows():
        file_name = '/'.join(row['path'].split('/')[1:])
        pattern = glob.glob(os.path.join(source_dir, file_name, '*'))
        
        for file in pattern:
            sub_folder = file.split('/')[-1]
            source_path = file
            dest_file_path = os.path.join(source_dir, set_data, file_name, sub_folder)
            os.makedirs(os.path.dirname(dest_file_path), exist_ok = True)
            
            if os.path.exists(source_path):
                shutil.move(source_path, dest_file_path)
        
        
```

```python
move_files(train_data, origin+'complete_dataset', set_data='Train')
```

```python
move_files(test_data, origin+'complete_dataset', set_data='Test')
```

```python
move_files(val_data, origin+'complete_dataset', set_data='Val')
```
